```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

global_ev_sales = pd.read_csv("global_ev_sales_2016_2023.csv")
tesla_sales = pd.read_csv("tesla_sales_2016_2023.csv")
tesla_market_share = pd.read_csv("tesla_market_share_2016_2023.csv")

df = global_ev_sales.merge(tesla_sales, on='Year').merge(tesla_market_share, on='Year')

print("Merged Dataset Preview:")
print(df.head())
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=df['Year'], y=df['Global EV Sales (millions)'], color='lightblue', label='Global EV Sales')
sns.barplot(x=df['Year'], y=df['Tesla Sales (units)']/1e6, color='red', label='Tesla Sales')
plt.xlabel('Year')
plt.ylabel('Sales (millions)')
plt.title('Tesla vs. Global EV Sales (2016-2023)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Year'], y=df['Tesla Market Share (%)'], marker='o', color='green')
plt.xlabel('Year')
plt.ylabel('Tesla Market Share (%)')
plt.title("Tesla's Market Share in Global EV Industry (2016-2023)")
plt.grid(True)
plt.show()
```

```python
print("\nKey Insights:")
print("1️⃣ Tesla's EV sales have grown significantly, but the global EV market is expanding even faster.")
print("2️⃣ Tesla’s market share peaked around 2020 but has declined due to increasing competition from BYD, VW, and other automakers.")
print("3️⃣ Tesla’s future growth will depend on innovation, production scaling, and maintaining competitive pricing amid rising global EV adoption.")
```
