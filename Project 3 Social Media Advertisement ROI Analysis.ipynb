```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("ad_campaign_data.csv")


df.drop_duplicates(inplace=True)

df.fillna(df.mean(), inplace=True)


df.columns = df.columns.str.lower().str.replace(" ", "_")

df['roi'] = ((df['revenue'] - df['ad_spend']) / df['ad_spend']) * 100

print(df[['campaign_id', 'platform', 'roi']].sort_values(by='roi', ascending=False).head())

corr_matrix = df[['ad_spend', 'clicks', 'conversions', 'revenue']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

sns.barplot(x='platform', y='roi', data=df)
plt.title("ROI Comparison by Platform")
plt.show()

df.groupby('campaign_id')[['ad_spend', 'revenue']].sum().plot(kind='line')
plt.title("Ad Spend vs Revenue Over Time")
plt.show()

sns.scatterplot(x='clicks', y='conversions', data=df)
plt.title("Clicks vs Conversions")
plt.show()
```
